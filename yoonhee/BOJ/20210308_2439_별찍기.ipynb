# 별찍기
첫째 줄에는 별 1개, 둘째 줄에는 별 2개, N번째 줄에는 별 N개를 찍는 문제

하지만, 오른쪽을 기준으로 정렬한 별(예제 참고)을 출력하시오.

```python
lines = int(input())
for line in range(lines):
    line += 1
    blanks = lines - line
    
    for _ in range(blanks) :
        print(" ", end='')
    for _ in range(line):
        print("*", end='')
    print()    
```
